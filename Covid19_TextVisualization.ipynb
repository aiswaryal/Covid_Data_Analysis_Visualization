```
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
```

```
df1 = pd.read_csv('covid+death.csv')
df1.head()
```

```
# Creating a WordCloud for condition
#Step 1 :COnvert the column with diseases count into a list using tolist()
sentences = df1['Condition'].tolist()
```

```
#Step 2 : Convert the list into one simple string
sentencesAsString = ' '.join(sentences)
```

```
#Step3 : Convert the string into wordcloud
plt.figure(figsize=(14,14))
plt.imshow(WordCloud().generate(sentencesAsString))
```

```
#creating a wordcloud for condition group
sent = df1['Condition Group'].tolist()
```

```
sentAsString = ' '.join(sent)
```

```
plt.figure(figsize=(20,20))
plt.imshow(WordCloud().generate(sentAsString))
```
